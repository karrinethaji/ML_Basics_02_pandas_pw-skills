```python
# Merge and Join

# Let's see Merge Operations: Merge is nothing but merging the data based on the matching of column element.
```

```python
data1=pd.DataFrame({"key1":[11,22,33,44],
                   "key2":[55,66,77,88],
                   "key3":[99,100,110,120]})
```

```python
data2=pd.DataFrame({"key1":[11,22,33,62],
                   "key4":[55,78,77,88],
                   "key5":[99,100,97,120]})
```

```python
data1
```

```python
data2
```


pd.merge(

    left: 'DataFrame | Series',
    right: 'DataFrame | Series',
    
    how: 'MergeHow' = 'inner',
    
    on: 'IndexLabel | None' = None,
    left_on: 'IndexLabel | None' = None,
    right_on: 'IndexLabel | None' = None,
    left_index: 'bool' = False,
    right_index: 'bool' = False,
    sort: 'bool' = False,
    suffixes: 'Suffixes' = ('_x', '_y'),
    copy: 'bool | None' = None,
    indicator: 'str | bool' = False,
    validate: 'str | None' = None,
) -> 'DataFrame'

```python
pd.merge(data1,data2)

# By default the join is inner join

# In data1 and data2 the key1 column is matching so, inner join will check the condition in the key1 and whenever the element in 
# key1 column from the both the dataframes are matching then it will be shown under the output along with the other columns data 
```

```python
# from the above output we can conclude that the elements: 11,22,33 in the key1 column is matching with the two dataframes at the exact
# indexes , so the elements printed in the output.
```

```python
data1
```

```python
data2
```

how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'
    Type of merge to be performed.

    * left: use only keys from left frame, similar to a SQL left outer join;
      preserve key order.
    * right: use only keys from right frame, similar to a SQL right outer join;
      preserve key order.
    * outer: use union of keys from both frames, similar to a SQL full outer
      join; sort keys lexicographically.
    * inner: use intersection of keys from both frames, similar to a SQL inner
      join; preserve the order of the left keys.
    * cross: creates the cartesian product from both frames, preserves the order
      of the left keys.

```python
# Outer join

pd.merge(data1,data2,how="outer")
```

```python
# Left join : This is the by default join

pd.merge(data1,data2,how="inner")
```

```python
# Right Join

pd.merge(data1,data2,how="right")
```

```python
# Cross Join: creates the cartesian product from both frames, preserves the order of the left keys.

pd.merge(data1,data2,how="cross")
```

Concept of Joins: https://www.geeksforgeeks.org/sql-join-set-1-inner-left-right-and-full-joins/

```python
# By default joins merge module will take the matching column 

# If you want to perform the join on the specific column i,e if there are two matching columns in two dataframes then, you can
# use 2nd column as well to perform the joins using the "on" parameter
```

```python
data3=pd.DataFrame({"key1":[11,22,33,44],
                   "key2":[55,66,77,88],
                   "key3":[99,100,110,120]})
```

```python
data4=pd.DataFrame({"key1":[11,22,33,62],
                   "key2":[55,78,77,88],
                   "key5":[99,100,97,120]})
```

```python
pd.merge(data3,data4,how="inner",on="key2")
```

```python
pd.merge(data3,data4,how="inner",on="key1")
```

```python
data3=pd.DataFrame({"key1":[11,22,33,44],
                   "key2":[55,66,77,88],
                   "key3":[99,100,110,120]})
```

```python
data4=pd.DataFrame({"key1":[88,45,56,99],
                   "key2":[11,22,56,44],
                   "key5":[99,100,97,120]})
```

```python
# If you want to check the condition from two different columns then 

pd.merge(data3,data4,how="inner",left_on="key1",right_on="key2")

# if there is any element matching with the key1 column from data3 df and key2 column from data4 df, then it will work 
# according to join specified,
```

```python
pd.merge(data3,data4,how="outer",left_on="key1",right_on="key2")
```

```python
pd.merge(data3,data4,how="left",left_on="key1",right_on="key2")
```

```python
# Joins: Join is also same as merge operation, but here the joining is based on the index values not based on the column elements
```

```python
data1=pd.DataFrame({"key1":[1,2,4,5,6],
                   "key2":[4,5,6,7,8],
                   "key3":[3,4,5,6,6]},index=list("abcde"))
```

```python
data1
```

```python
data2=pd.DataFrame({"key11":[1,2,4,5,6],
                   "key22":[4,5,6,7,8],
                   "key33":[3,4,5,6,6]},index=["a","b","h","i","j"])
```

```python
data2
```

```python
data1.join(data2)

# by default join is "Left join"

# Left table is data1 and right table is data2 

# Here indexes a and b only matching so we are getting the complete row details for those indexes and for remaining indexes

# we are getting the right table values as we are using "left join"
```

```python
data1.join(data2,how="left") # the below output exactly same as above output
```

```python
data1.join(data2,how="inner")
```

```python
data1.join(data2,how="outer") # outer means all the indexes will be compared from the two df 
```

```python
data1.join(data2,how="right") # indexes from the right table will be compared with the left table
```

```python
data1.join(data2,how="cross")
```
